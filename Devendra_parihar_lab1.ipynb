```python
# importing the dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
# uploading the dataset
from google.colab import files
files.upload()
```

```python
# import dataset to Pandas Data Frame
rainfall = pd.read_csv('rainfall in india 1901-2015.csv') 
#storing the data frame to the variable 'rainfall'
```

```python
rainfall.head()
```

```python
rainfall = rainfall.rename(columns = {'SUBDIVISION':'state', 'JAN': 'jan', 'FEB': 'feb', 'MAR': 'mar', 
                       'APR': 'apr' , 'MAY': 'may', 'JUN': 'jun', 'JUL': 'jul', 'AUG': 'aug', 'SEP': 'sep', 'OCT': 'oct'
                      , 'NOV': 'nov', 'DEC': 'dec', 'ANNUAL': 'annual', 'YEAR': 'year'})
rainfall.tail()#We changed the label of the rows as per my ease of operation
```

```python
rainfall.state.unique()
```

```python
rainfall.year.unique()
```

```python
rainfall.groupby("year").sum()['annual'].plot(figsize=(12,8));
```

```python
#We created two specific data set one from 
#KONKAN, and other for Kerla to process the data in further codes
rainfall_Raj = rainfall[rainfall.state=='EAST RAJASTHAN']
rainfall_kel = rainfall[rainfall.state=='KERALA']
```

```python
#BAR PLOT
plt.figure(figsize = (20,5))
sns.barplot(x='year', y= 'annual', data = rainfall_Raj)
plt.xticks(rotation = 90)
plt.title('East Rajasthan RAINFALL DATA FROM 1901-2015')
plt.show()
```

```python
#BAR PLOT
plt.figure(figsize = (20,5))
sns.barplot(x='year', y= 'annual', data = rainfall_kel)
plt.xticks(rotation = 90)
plt.title('Kerla RAINFALL DATA FROM 1901-2015')
plt.show()
```

```python
#Line plot
plt.figure(figsize=(25,15))
sns.lineplot(x = 'year', y= 'annual', hue = 'state', data = rainfall)
```

```python
rainfall_Raj.groupby(['year'])['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'].sum().plot.line(figsize=(15,5))
plt.title('East Rajasthan')
rainfall_kel.groupby(['year'])['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'].sum().plot.line(figsize=(15,5))
plt.title('KERALA')
plt.show()
```

Above charts shows the rainfall data for two states

```python
#here we used groupby function and sorted the values in ascending order to create rainfall data
plt.figure(figsize=(15,5))
rainfall.groupby(['state','year'])['annual'].sum().sort_values(ascending=False).plot()
plt.xticks(rotation=90)
```

```python
#Histogram
plt.figure(figsize=(15,5))
rainfall.groupby(['year'])['annual'].sum().sort_values(ascending=False).head(12).plot(kind='hist', color = 'b')
plt.ylabel('Yearly Rainfall')
plt.title('Yearly Rainfall Data')
```

```python
#Bar plot
plt.figure(figsize=(15,5))
rainfall.groupby(['year'])['annual'].sum().sort_values(ascending=False).head(12).plot(kind='bar', color = 'r')
plt.ylabel('Yearly Rainfall')
plt.title('Yearly Rainfall Data')
```

```python
#Line plot
plt.figure(figsize=(15,5))
rainfall.groupby(['year'])['annual'].sum().sort_values(ascending=False).head(12).plot(kind='line', color = 'g')
plt.ylabel('Yearly Rainfall')
plt.title('Yearly Rainfall Data')
```

```python
plt.figure(figsize=(15,5))
rainfall.groupby(['state'])['annual'].sum().sort_values(ascending=False).head(12).plot(kind='bar', color = 'y')
plt.ylabel('Total Rainfall')
plt.title('Total Rainfall Data')
```

```python
plt.figure(figsize=(10,5))
rainfall[['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug',
       'sep', 'oct', 'nov', 'dec']].mean().plot(kind= 'bar')
plt.xlabel('Months')
plt.ylabel('Avg. Rainfall')
plt.title('Avg. Monthly Rainfall Data')
plt.show()
```
